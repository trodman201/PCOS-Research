```python
import pandas as pd
```

```python
df = pd.read_excel('PCOS_data_without_infertility.xlsx', sheet_name='Full_new')
df
```

```python
new_df = df[['PCOS (Y/N)','Weight (Kg)', 'BMI','FSH/LH']]
new_df
```

```python
pcos = new_df[new_df['PCOS (Y/N)'].isin([1])]
pcos
```

```python
no_pcos = new_df[new_df['PCOS (Y/N)'].isin([0])]
```

```python
p1 = new_df[new_df['PCOS (Y/N)'].isin([1])]
p1 = p1[p1['BMI'] <= 19]
p1
```

```python
p2 = new_df[new_df['PCOS (Y/N)'].isin([1])]
p2 = p2[(p2['BMI'] >= 19) & (p2['BMI'] <= 25)]
p2
```

```python
p3 = new_df[new_df['PCOS (Y/N)'].isin([1])]
# Drop rows where 'BMI' is less than 25
p3 = p3[(p3['BMI'] >= 25)]
p3
p3.dropna()
```

```python
#underweight no pcos
n1 = new_df[new_df['PCOS (Y/N)'].isin([0])]
n1 = n1[n1['BMI'] <= 19]
```

```python
n2 = new_df[new_df['PCOS (Y/N)'].isin([0])]
n2 = n2[(n2['BMI'] >= 19) & (n2['BMI'] <= 25)]
n2
```

```python
n3 = new_df[new_df['PCOS (Y/N)'].isin([0])]
# Drop rows where 'BMI' is less than 25
n3 = n3[(n3['BMI'] >= 25)]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x="PCOS (Y/N)", y="BMI", data=p1)
sns.boxplot(x="PCOS (Y/N)", y="BMI", data=n1)
plt.title('Underweight Groups')
```

```python
sns.boxplot(x="PCOS (Y/N)", y="BMI", data=p2)
sns.boxplot(x="PCOS (Y/N)", y="BMI", data=n2)
plt.title('Healthy weight Groups')
```

```python
sns.boxplot(x="PCOS (Y/N)", y="BMI", data=p3)
sns.boxplot(x="PCOS (Y/N)", y="BMI", data=n3)
plt.title('Overweight Groups')
```

```python
sns.boxplot(x="PCOS (Y/N)", y="BMI", data=pcos)
sns.boxplot(x="PCOS (Y/N)", y="BMI", data=no_pcos)
plt.title('PCOS VS No PCOS by BMI')
```

```python
sns.boxplot(x="PCOS (Y/N)", y="FSH/LH", data=p1)
plt.title('Underweight PCOS Group and their FSH/LH')
```

```python
sns.boxplot(x="PCOS (Y/N)", y="FSH/LH", data=p2)
plt.title('Normal Weight PCOS Group and their FSH/LH')
```

```python
print("Unique values in 'FSH/LH' column of p3 before filtering:")
print(p3['FSH/LH'].unique())
values_to_drop = [327, 61.875, 21.33333333]
filtered_p3 = p3[~p3['FSH/LH'].isin(values_to_drop)]
print("Unique values in 'FSH/LH' column of p3 after filtering:")
print(filtered_p3['FSH/LH'].unique())
```

```python
sns.boxplot(x="PCOS (Y/N)", y="FSH/LH", data=filtered_p3)
plt.title('Overweight PCOS Group and their FSH Levels')
```

```python
sns.boxplot(x="PCOS (Y/N)", y="FSH/LH", data=n1)
plt.title('Underweight non-PCOS Group and their FSH/LH')
```

```python
threshold_value = 20
filtered_n2 = n2[n2['FSH/LH'] <= threshold_value]

# Create boxplot with the filtered data
sns.boxplot(x="PCOS (Y/N)", y="FSH/LH", data=filtered_n2)

plt.title('Normal non-PCOS Group and their FSH/LH')
plt.show()
```

```python
sns.boxplot(x="PCOS (Y/N)", y="FSH/LH", data=n3)
plt.title('Overweight  non-PCOS Group and their FSH/LH')
```
